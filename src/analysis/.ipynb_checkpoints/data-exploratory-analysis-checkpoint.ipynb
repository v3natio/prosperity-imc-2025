```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
df = pd.read_csv("/home/venatio/documents/projects/prosperity-imc-2025/data/round_1/prices_round_1_day_-2.csv", sep=';')

# Show the first few rows of the data
df.head()
```

```python
# Filter by a specific product
product_name = "KELP"  # Change this to "KELP" or "SQUID_INK" if needed
filtered_df = df[df['product'] == product_name]

# Plot mid_price (x) vs. timestamp (y)
plt.figure(figsize=(40, 12))
plt.plot(filtered_df['timestamp'], filtered_df['mid_price'], linestyle='-')

plt.title(f"{product_name}")
plt.xlabel("Timestamp")
plt.ylabel("Mid price")
plt.grid(True)
plt.show()
```
