```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the CSV files
df_2 = pd.read_csv("../../data/round_1/prices_round_1_day_-2.csv", sep=';')
df_1 = pd.read_csv("../../data/round_1/prices_round_1_day_-1.csv", sep=';')
df_0 = pd.read_csv("../../data/round_1/prices_round_1_day_0.csv", sep=';')

merged_df = pd.concat([df_2, df_1, df_0], ignore_index=True)
```

```python
merged_df.head()
```

```python
# Filter by a specific product
product_name = "SQUID_INK"
filtered_df_2 = df_2[df_2['product'] == product_name]
```

```python
# Plot mid_price (x) vs. timestamp (y)
plt.figure(figsize=(40, 12))
plt.plot(filtered_df_2['timestamp'], filtered_df_2['mid_price'], linestyle='-')

plt.title(f"{product_name}")
plt.xlabel("Timestamp")
plt.ylabel("Mid price")
plt.grid(True)
plt.show()
```
